```python
import numpy as np
import random
from collections import deque
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
class Queens():
    def __init__(self, n):
        self.size = n
        self.board = np.zeros((n, n), dtype=int)
        self.reset()

    def put(self, x, y):
        self.board[x][y] = 1


    def reset(self):
        self.board = np.zeros((self.size, self.size), dtype=int)
        return self.board
```

```python
def queencheck(board):
    b1 = np.sum(board, axis=0)
    b2 = np.sum(board, axis=1)
    c1 = np.all(b1 <= 1)
    c2 = np.all(b2 <= 1)
    for d in range(-len(board) + 1, len(board)):
        if np.sum(np.diagonal(board, offset=d)) > 1:
            return False
    flipped_board = np.fliplr(board)
    for d in range(-len(flipped_board) + 1, len(flipped_board)):
        if np.sum(np.diagonal(flipped_board, offset=d)) > 1:
            return False
    return c1 and c2
```

```python
class DQNAgent(nn.Module):
    def __init__(self, n):
        super(DQNAgent, self).__init__()
        self.fc1 = nn.Linear(n*n, n*n)
        self.fc2 = nn.Linear(n*n, n*n)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def trainer(episodes=500, n=4, temperature=1.0):
    env = Queens(n)  # n-퀸 환경 초기화
    agent = DQNAgent(n)  # DQN 에이전트 초기화
    optimizer = optim.Adam(agent.parameters(), lr=0.1)  # 최적화 알고리즘 설정
    memory = deque(maxlen=10000)  # 경험 재생 메모리 초기화
    batch_size = 1000  # 배치 크기 설정
    gamma = 0.8  # 미래 보상에 대한 할인 계수 설정
    rewards_per_episode = []

    for episode in range(episodes):  # 에피소드 반복
        state = env.reset()  # 환경을 초기 상태로 리셋
        state_flat = torch.FloatTensor(state.flatten()).unsqueeze(0)  # 상태를 평탄화하고 텐서로 변환
        total_reward = 0  # 총 보상 초기화
        step_reward = 1  # 초기 스텝 보상 설정

        for step in range(n):  # 한 에피소드 내의 각 스텝 반복
            with torch.no_grad():  # 기울기 계산을 중지하여 메모리 사용량과 연산 시간 절약
                q_values = agent(state_flat)  # 현재 상태에 대한 Q값 예측

            # 중복 행동 방지를 위해 이미 선택된 행과 열의 위치에 대한 Q값 마스킹
            masked_q_values = q_values.clone()
            for row in range(n):
                if 1 in env.board[row]:  # 해당 행에 퀸이 이미 배치되어 있는 경우
                    masked_q_values[0, row*n:(row+1)*n] = float('-inf')  # 해당 행 전체를 마스킹

            for col in range(n):
                if 1 in env.board[:,col]:  # 해당 열에 퀸이 이미 배치되어 있는 경우
                    masked_q_values[0, col::n] = float('-inf')  # 해당 열 전체를 마스킹

            # 온도 스케일링을 사용한 행동 선택
            probabilities = F.softmax(masked_q_values / temperature, dim=1)
            action = np.random.choice(np.arange(n * n), p=probabilities.numpy().ravel())

            x, y = divmod(action, n)  # 행동을 좌표로 변환
            if env.board[x][y] == 0:  # 유효한 행동인 경우
                env.put(x, y)  # 퀸 배치
                if queencheck(env.board):  # 보드 상태가 여전히 유효한 경우
                    reward = step_reward  # 현재 스텝에서의 보상
                    step_reward *= 3  # 다음 스텝의 보상을 두 배로 증가
                else:  # 보드 상태가 유효하지 않은 경우
                    reward = -4  # 유효하지 않은 행동에 대한 패널티
                    step_reward = 1  # 스텝 보상을 초기화
            else:  # 유효하지 않은 행동에 대한 패널티
                reward = -10
                step_reward = 1  # 스텝 보상을 초기화

            total_reward += reward  # 총 보상 업데이트
            # 다음 상태 준비
            next_state_flat = torch.FloatTensor(env.board.flatten()).unsqueeze(0)
            memory.append((state_flat, action, reward, next_state_flat))  # 메모리에 경험 저장

            state_flat = next_state_flat  # 현재 상태 업데이트

            # 학습
            if len(memory) > batch_size:  # 메모리에 충분한 경험이 쌓인 경우
                batch = random.sample(memory, batch_size)  # 배치 샘플링
                states, actions, rewards, next_states = zip(*batch)  # 배치 데이터 분할

                states = torch.cat(states)
                actions = torch.tensor(actions)
                rewards = torch.tensor(rewards, dtype=torch.float32)
                next_states = torch.cat(next_states)

                # 선택한 행동에 대한 Q값 예측
                q_values = agent(states).gather(1, actions.unsqueeze(1)).squeeze(1)

                # 다음 상태에 대한 최대 Q값 예측
                next_q_values = agent(next_states).max(1)[0]

                # 기대 Q값 계산
                expected_q_values = rewards + gamma * next_q_values

                # 손실 계산 및 역전파
                loss = F.mse_loss(q_values, expected_q_values)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

        # 각 에피소드의 마지막에 보드 상태와 총 보상 출력
        print(f"Episode {episode+1} completed.\nBoard:\n{env.board}\nTotal Reward: {total_reward}\n")
        rewards_per_episode.append(total_reward)
    return rewards_per_episode
```

```python
rewards = trainer(episodes=25000, n=10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(rewards, label='Total Rewards per Episode')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Rewards per Episode Over Time')
plt.legend()
plt.show()
```
